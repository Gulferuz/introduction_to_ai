Создаем массивы NumPy

```python
arr_1 = np.array([1, 2, 3])
arr_2 = np.array([4, 5, 6])
print(arr_1 + arr_2) # Вывод: [5 7 9] (1+4, 2+5, 3+6)
```

### Давайте решим задачу линейной регресий

  - среднее значение 

```python
# Исходные данные (обычные списки)
x_list = [1, 2, 3, 4]
y_list = [3, 5, 7, 9]
n = len(x_list)

# Преобразуем в массивы NumPy
x = np.array(x_list)
y = np.array(y_list)
```

Шаг 1: Вычисляем средние значения (x̄ и ȳ)

```python
x_mean = sum(x) / n
y_mean = sum(y) / n

print(f"Среднее x (x̄) = {x_mean}")
print(f"Среднее y (ȳ) = {y_mean}")
```

---
Шаг 2: Вычисляем числитель и знаменатель для w₁
 
Числитель: Σ(xᵢ - x̄)(yᵢ - ȳ)
Знаменатель: Σ(xᵢ - x̄)²

_np.sum() — это функция для вычисления суммы всех элементов массива_

```python
# NumPy сам вычтет x_mean из каждого элемента массива x
numerator = np.sum((x - x_mean) * (y - y_mean))

# И сам возведет каждое отклонение в квадрат перед суммированием
denominator = np.sum((x - x_mean)**2)

```

---
Шаг 3: Вычисляем w₁ (коэффициент наклона)

```python
w1 = numerator / denominator
```

```python
w0 = y_mean - w1 * x_mean
```

```python
print(f"w₀ = {w0}, w₁ = {w1}")
print(f"Модель: y = {w0} + {w1}x")
```

Финальная версия для линейной регрессий в numpy

```python
import numpy as np 

x_list = [1, 2, 3, 4]
y_list = [3, 5, 7, 9]
n = len(x_list)

x = np.array(x_list)
y = np.array(y_list)

x_mean = sum(x) / n
y_mean = sum(y) / n

print(f"Среднее x (x̄) = {x_mean}")
print(f"Среднее y (ȳ) = {y_mean}")

numerator = np.sum((x - x_mean) * (y - y_mean))

denominator = np.sum((x - x_mean)**2)

w1 = numerator / denominator

w0 = y_mean - w1 * x_mean

print(f"w₀ = {w0}, w₁ = {w1}")
print(f"Модель: y = {w0} + {w1}x")
```

</div>

это самая популярная библиотека для классического машинного обучения на Python.

Если NumPy — это калькулятор, то scikit-learn — это целый завод. Вам не нужно прописывать формулы числителя и знаменателя. Вы просто «скармливаете» данные модели, и она сама подбирает коэффициенты $w_1$ и $w_0$.

Главные плюсы:

Скорость: Алгоритмы оптимизированы и работают молниеносно.

Единообразие: Все модели (линейная регрессия, нейросети, деревья решений) обучаются одинаково — через команду .fit().

_pip install scikit-learn_  - Установка

```python
from sklearn.linear_model import LinearRegression 
# Импортируем класс LinearRegression из модуля linear_model библиотеки scikit-learn
```

```python
model = LinearRegression() # Эта строка создает модель линейной регрессии. 
```

_model_ - это переменная, в которой хранится наша модель

_LinearRegression()_ - это конструктор (специальная функция), которая создает пустую модель
Пока модель не обучена и не знает никаких закономерностей

```python
X = np.array([[1], [2], [3] ,[4]])
y = np.array([3, 5, 7, 9])
print(f"X={X} \n y={y}")
```

Теперь нужно ОБУЧИТЬ модель методом .fit()

model.fit(X, y) - это метод, который подбирает коэффициенты w0 и w1.

X - признаки (входные данные)

y - целевые значения (правильные ответы)

Во время fit() модель "смотрит" на данные и находит закономерность y = w0 + w1*x

model.fit(X, y) - эту строку нужно добавить после создания модели!


```python
model.fit(X, y);
```

После обучения можно делать ПРЕДСКАЗАНИЯ методом _.predict()_

```python
print(model.predict([[20]])) 
```

Финальная версия для scikit-learn

```python
import numpy as np 
from sklearn.linear_model import LinearRegression 

model = LinearRegression()
X = np.array([[1], [2], [3] ,[4]])
y = np.array([3, 5, 7, 9])
print(f"X={X} \n y={y}")
model.fit(X, y);

print(model.predict([[20]])) 
```

## Объединяем NumPy, Matplotlib и scikit-learn

np.linspace(0, 5, 100) - это

_Мы создаем 100 точек, равномерно распределенных от 0 до 5. Это наша «виртуальная линейка»._ 


.reshape(-1, 1) - это

_Библиотека sklearn работает только с таблицами (столбцами). Эта команда берет плоский список из 100 чисел и выстраивает их в один вертикальный столбик._

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([[1], [2], [3], [4]]) 
y = np.array([3, 5, 7, 9])

model = LinearRegression()
model.fit(x, y) 

w1 = model.coef_[0]    
w0 = model.intercept_  
print(f"Подобранная формула: y = {w0} + {w1}*x")


x_range = np.linspace(0, 5, 100).reshape(-1, 1) 
y_pred = model.predict(x_range)                

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Реальные данные') 
plt.plot(x_range, y_pred, color='red', label='Линия sklearn')
plt.title('Линейная регрессия через scikit-learn')
plt.xlabel('Картошка')
plt.ylabel('Бег (км)')
plt.legend()
plt.grid(True)
plt.show()

print(model.predict([[40]])) 
```
